```python
for col in df.columns:
    mode = df[col].mode(dropna=True)
    if not mode.empty:
        df[col].fillna(mode[0], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.drop(['capital-gain','capital-loss'],axis=1)
```

```python
numeric_cols = df.select_dtypes(include=['int', 'float']).columns

df_capped = df.copy()

for col in numeric_cols:
    Q1 = df_capped[col].quantile(0.25)
    Q3 = df_capped[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Cap outliers to the IQR range
    df_capped[col] = df_capped[col].clip(lower=lower, upper=upper)

print(f"Original shape: {df.shape}")
print(f"Shape after winsorization: {df_capped.shape}")  # Will be the same!
df = df_capped
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
label_mappings = {}  # Dictionary to hold mapping for each column

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    # Store mapping: number -> category
    label_mappings[col] = dict(zip(le.transform(le.classes_), le.classes_))

# To display all mappings:
for col, mapping in label_mappings.items():
    print(f"\nColumn: {col}")
    for num, category in mapping.items():
        print(f"  {num}: {category}")
```

```python
def print_section(title, emoji=""):
    print("\n", "="*70)
    print(f"{emoji} {title}")
    print("","="*70)
```

```python
def objective(trial):
    valid_config = [
        ('liblinear','l1'),
        ('liblinear','l2'),
        ('newton-cg','l2'),
        ('saga','l1'),
        ('saga','l2'),
        ('lbfgs','l2')
    ]

    config = trial.suggest_categorical("solver_penalty",valid_config)
    solver, penalty = config

    params= {
        "solver":solver,
        "penalty":penalty,
        "dual":False,
        "C":trial.suggest_float("C",0.1,1.5,log=True),
        "max_iter": trial.suggest_int("max_iter",10,120),
        "fit_intercept": trial.suggest_categorical("fit_intercept",[True, False]),
        "warm_start": trial.suggest_categorical("warm_start",[True, False]),
        "class_weight": trial.suggest_categorical("class_weight",[None, "balanced"]),
        "intercept_scaling": trial.suggest_float("intercept_scaling",0.01,1.5) if solver in ["liblinear","saga"] else 1.0
    }

    model = LogisticRegression(**params)
    score = cross_val_score(model,X_train, y_train,cv=5, scoring="accuracy").mean()

    print(f"\nTrial: {trial.number} Score:{score:.4f}")
    print("Params",params)
    return score
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)
```

```python
print_section("Evaluation of Logistic Regression", emoji="üîç")

print_section("Logistic Regression WITHOUT Optuna Hyperparameters", emoji="üß©")
model_lr = LogisticRegression()
model_lr.fit(X_train,y_train)
pred = model_lr.predict(X_test)

acc_lr = accuracy_score(pred, y_test)
prec = precision_score(pred, y_test)
rec = recall_score(pred, y_test)
f1 = f1_score(pred, y_test)

print(f"üîπ Accuracy Score: {acc_lr:.4f}")
print(f"üîπ Precision Score: {prec:.4f}")
print(f"üîπ Recall Score: {rec:.4f}")
print(f"üîπ F1 Score: {f1:.4f}")

print("Model HyperParameters Used: ")
params_df = pd.DataFrame(model_lr.get_params(),index=['Value']).T
print(params_df)

print("\nConfusion Matrix:")
print(pd.DataFrame(confusion_matrix(y_test, pred)))

print("\nClassification Report:")
print(classification_report(y_test, pred, zero_division=0))

print_section("Logistic Regression WITH Optuna Hyperparameters", emoji="‚ú®")
best_params = study.best_trial.params.copy()
# Unpack the solver_penalty if used as a tuple
if "solver_penalty" in best_params:
    best_params["solver"], best_params["penalty"] = best_params["solver_penalty"]
    del best_params["solver_penalty"]

model_lr_optuna = LogisticRegression(**best_params)
model_lr_optuna.fit(X_train, y_train)
pred_optuna = model_lr_optuna.predict(X_test)

acc_lr_opt = accuracy_score(y_test, pred_optuna)
prec_opt = precision_score(y_test, pred_optuna, average='weighted', zero_division=0)
rec_opt = recall_score(y_test, pred_optuna, average='weighted', zero_division=0)
f1_opt = f1_score(y_test, pred_optuna, average='weighted', zero_division=0)

print(f"\nüîπ Accuracy Score (Optuna):      {acc_lr_opt:.4f}")
print(f"üîπ Precision (weighted):         {prec_opt:.4f}")
print(f"üîπ Recall (weighted):            {rec_opt:.4f}")
print(f"üîπ F1 Score (weighted):          {f1_opt:.4f}\n")

print(f"üéØ Best Optuna Trial Accuracy (CV): {study.best_trial.value:.4f}")
print("\nBest Optuna Hyperparameters:")
params_df_optuna = pd.DataFrame(model_lr_optuna.get_params(), index=["Value"]).T
print(params_df_optuna)

print("Confusion Matrix (Optuna):")
print(pd.DataFrame(confusion_matrix(y_test, pred_optuna)))

print("\nClassification Report (Optuna):")
print(classification_report(y_test, pred_optuna, zero_division=0))
```

```python
def objective_svc(trial):
    params = {
        "C": trial.suggest_float("C", 0.1, 15.0, log=True),
        "kernel": trial.suggest_categorical("kernel", ["linear", "poly", "rbf", "sigmoid"]),
        "degree": trial.suggest_int("degree", 2, 5) if trial.suggest_categorical("use_poly", [True, False]) else 3,
        "gamma": trial.suggest_categorical("gamma", ["scale", "auto"]),
        "coef0": trial.suggest_float("coef0", 0.0, 3.0) if trial.suggest_categorical("use_poly", [True, False]) else 0.0,
        "shrinking": trial.suggest_categorical("shrinking", [True, False]),
        "probability": trial.suggest_categorical("probability", [False, True]),
        "tol": trial.suggest_float("tol", 1e-4, 1e-1, log=True),
        "class_weight": trial.suggest_categorical("class_weight", [None, "balanced"]),
        # "max_iter": -1,   # Use default for now; can include for very large datasets
    }
    # Remove unnecessary keys if not using poly kernel
    if params["kernel"] != "poly":
        params["degree"] = 3
        params["coef0"] = 0.0
    model = SVC(**params)
    score = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy", n_jobs=-1).mean()
    return score

#Run Optuna Optimization
study_svc = optuna.create_study(direction="maximize")
study_svc.optimize(objective_svc, n_trials=3)
```

```python
def print_section(title, emoji=""):
    print("\n" + "="*70)
    print(f"{emoji} {title}")
    print("="*70 + "\n")

# Evaluation with Default Parameters
print_section("Evaluation of Support Vector Classifier", emoji="üî∑")

print_section("SVC WITHOUT Optuna Hyperparameters", emoji="üß©")
model_svc = SVC()
model_svc.fit(X_train, y_train)
pred_svc = model_svc.predict(X_test)

acc_svc = accuracy_score(y_test, pred_svc)
prec = precision_score(y_test, pred_svc, average='weighted', zero_division=0)
rec = recall_score(y_test, pred_svc, average='weighted', zero_division=0)
f1 = f1_score(y_test, pred_svc, average='weighted', zero_division=0)

print(f"üîπ Accuracy Score:      {acc_svc:.4f}")
print(f"üîπ Precision (weighted): {prec:.4f}")
print(f"üîπ Recall (weighted):    {rec:.4f}")
print(f"üîπ F1 Score (weighted):  {f1:.4f}\n")

print("Model Hyperparameters Used:")
params_df = pd.DataFrame(model_svc.get_params(), index=["Value"]).T
print(params_df)

print("\nConfusion Matrix:")
print(pd.DataFrame(confusion_matrix(y_test, pred_svc)))

print("\nClassification Report:")
print(classification_report(y_test, pred_svc, zero_division=0))

# Optuna Objective Function for SVC
print_section("SVC WITH Optuna Hyperparameters", emoji="‚ú®")

best_params_svc = study_svc.best_trial.params.copy()
if "use_poly" in best_params_svc:
    del best_params_svc["use_poly"]
model_svc_optuna = SVC(**best_params_svc)
model_svc_optuna.fit(X_train, y_train)
pred_svc_optuna = model_svc_optuna.predict(X_test)

acc__svcopt = accuracy_score(y_test, pred_svc_optuna)
prec_opt = precision_score(y_test, pred_svc_optuna, average='weighted', zero_division=0)
rec_opt = recall_score(y_test, pred_svc_optuna, average='weighted', zero_division=0)
f1_opt = f1_score(y_test, pred_svc_optuna, average='weighted', zero_division=0)

print(f"üéØ Best Optuna Trial Accuracy (CV): {study_svc.best_trial.value:.4f}")
print("\nBest Optuna Hyperparameters:")
params_df_optuna = pd.DataFrame(best_params_svc, index=["Value"]).T
print(params_df_optuna)

print(f"\nüîπ Accuracy Score (Optuna):      {acc__svcopt:.4f}")
print(f"üîπ Precision (weighted):         {prec_opt:.4f}")
print(f"üîπ Recall (weighted):            {rec_opt:.4f}")
print(f"üîπ F1 Score (weighted):          {f1_opt:.4f}\n")

print("Confusion Matrix (Optuna):")
print(pd.DataFrame(confusion_matrix(y_test, pred_svc_optuna)))

print("\nClassification Report (Optuna):")
print(classification_report(y_test, pred_svc_optuna, zero_division=0))
```
